## Importing & Cleaning Data

```python
import pandas as pd
import numpy as np

df = pd.read_csv('totalprice.csv')
```

```python
pd.set_option('display.max_columns', 15)
pd.set_option('display.max_rows', 100)
```

```python
# Increasing readability by renaming columns
df.rename(columns={'Current Year Week Ending':'Date'}, inplace=True)
df.rename(columns={'ASP Current Year':'Price'}, inplace=True)
df.rename(columns={'Type':'Organic'}, inplace=True)

# Timeframe was redundant as it only had one value. 
df.drop(columns=['Timeframe'],inplace=True)

# Organic & its typo Conventional can be converted to Boolean.
df['Organic'] = df['Organic'].map({'Organic': True, 'Conventional': False, 'Conventional ': False})
df['Price'] = df['Price'].astype(float)

# Converting before new Index
df['Date'] = df['Date'].map(lambda x: x.rstrip(' 00:00:00'))
df['Date'] = pd.to_datetime(df['Date'], format= '%Y-%m-%d')
df.set_index('Date', inplace=True)
```

```python
# All locations were in the same column, creating new column allows easier groupby later.

# Total US
df.loc[df['Geography'] == 'Total U.S.', 'Regions'] = 'United States'
# Total Regions
df.loc[df['Geography'] == 'West', 'Regions'] = 'Total West'
df.loc[df['Geography'] == 'Plains', 'Regions'] = 'Total Plains'
df.loc[df['Geography'] == 'South Central', 'Regions'] = 'Total South Central'
df.loc[df['Geography'] == 'California', 'Regions'] = 'Total California'
df.loc[df['Geography'] == 'Great Lakes', 'Regions'] = 'Total Great Lakes'
df.loc[df['Geography'] == 'Northeast', 'Regions'] = 'Total Northeast'
df.loc[df['Geography'] == 'Midsouth', 'Regions'] = 'Total Midsouth'
df.loc[df['Geography'] == 'Southeast', 'Regions'] = 'Total Southeast'
# Total Cities
df.loc[df['Geography'] == 'San Diego', 'Regions'] = 'California'
df.loc[df['Geography'] == 'Los Angeles', 'Regions'] = 'California'
df.loc[df['Geography'] == 'San Francisco', 'Regions'] = 'California'
df.loc[df['Geography'] == 'Sacramento', 'Regions'] = 'California'

df.loc[df['Geography'] == 'Boise', 'Regions'] = 'West'
df.loc[df['Geography'] == 'Las Vegas', 'Regions'] = 'West'
df.loc[df['Geography'] == 'Spokane', 'Regions'] = 'West'
df.loc[df['Geography'] == 'Denver', 'Regions'] = 'West'
df.loc[df['Geography'] == 'Seattle', 'Regions'] = 'West'
df.loc[df['Geography'] == 'Phoenix/Tucson', 'Regions'] = 'West'
df.loc[df['Geography'] == 'Portland', 'Regions'] = 'West'
df.loc[df['Geography'] == 'West Tex/New Mexico', 'Regions'] = 'West'

df.loc[df['Geography'] == 'St. Louis', 'Regions'] = 'Plains'

df.loc[df['Geography'] == 'New Orleans/Mobile', 'Regions'] = 'South Central'
df.loc[df['Geography'] == 'Dallas/Ft. Worth', 'Regions'] = 'South Central'
df.loc[df['Geography'] == 'Houston', 'Regions'] = 'South Central'

df.loc[df['Geography'] == 'Indianapolis', 'Regions'] = 'Great Lakes'
df.loc[df['Geography'] == 'Cincinnati/Dayton', 'Regions'] = 'Great Lakes'
df.loc[df['Geography'] == 'Detroit', 'Regions'] = 'Great Lakes'
df.loc[df['Geography'] == 'Columbus', 'Regions'] = 'Great Lakes'
df.loc[df['Geography'] == 'Grand Rapids', 'Regions'] = 'Great Lakes'
df.loc[df['Geography'] == 'Chicago', 'Regions'] = 'Great Lakes'

df.loc[df['Geography'] == 'Harrisburg/Scranton', 'Regions'] = 'Northeast'
df.loc[df['Geography'] == 'Boston', 'Regions'] = 'Northeast'
df.loc[df['Geography'] == 'Syracuse', 'Regions'] = 'Northeast'
df.loc[df['Geography'] == 'Philadelphia', 'Regions'] = 'Northeast'
df.loc[df['Geography'] == 'Buffalo/Rochester', 'Regions'] = 'Northeast'
df.loc[df['Geography'] == 'Albany', 'Regions'] = 'Northeast'
df.loc[df['Geography'] == 'Pittsburgh', 'Regions'] = 'Northeast'
df.loc[df['Geography'] == 'New York', 'Regions'] = 'Northeast'
df.loc[df['Geography'] == 'Hartford/Springfield', 'Regions'] = 'Northeast'
df.loc[df['Geography'] == 'Northern New England', 'Regions'] = 'Northeast'

df.loc[df['Geography'] == 'Charlotte', 'Regions'] = 'Midsouth'
df.loc[df['Geography'] == 'Raleigh/Greensboro', 'Regions'] = 'Midsouth'
df.loc[df['Geography'] == 'Roanoke', 'Regions'] = 'Midsouth'
df.loc[df['Geography'] == 'Baltimore/Washington', 'Regions'] = 'Midsouth'
df.loc[df['Geography'] == 'Richmond/Norfolk', 'Regions'] = 'Midsouth'
df.loc[df['Geography'] == 'Louisville', 'Regions'] = 'Midsouth'
df.loc[df['Geography'] == 'South Carolina', 'Regions'] = 'Midsouth'
df.loc[df['Geography'] == 'Nashville', 'Regions'] = 'Midsouth'

df.loc[df['Geography'] == 'Tampa', 'Regions'] = 'Southeast'
df.loc[df['Geography'] == 'Orlando', 'Regions'] = 'Southeast'
df.loc[df['Geography'] == 'Atlanta', 'Regions'] = 'Southeast'
df.loc[df['Geography'] == 'Jacksonville', 'Regions'] = 'Southeast'
df.loc[df['Geography'] == 'Miami/Ft. Lauderdale', 'Regions'] = 'Southeast'
```

```python
# OT stands for organic priced rows.
final = df.copy()
OT = df.copy()
```

```python
# Seperating the Organic boolean value rows from each other.

OT = OT[OT['Organic'] == True]
final = final[final['Organic'] == False]
OT.rename(columns={'Price':'Organic Price'}, inplace=True)
final.rename(columns={'Price':'Regular Price'}, inplace=True)

OT.drop(columns=["4046 Units", "4225 Units","4770 Units","TotalBagged Units",
                 "SmlBagged Units","LrgBagged Units","X-LrgBagged Units"],inplace=True)
```

```python
# Concatenating
final['Organic Price'] = OT['Organic Price'].values
final.drop(columns=['Organic'],inplace=True)
final['Mean Price'] = final[['Regular Price', 'Organic Price']].mean(axis=1)
```

```python
# Organizing columns
cols = list(df.columns.values)
final = final[['Regions','Geography','Mean Price','Regular Price',
               'Organic Price','Total Bulk and Bags Units','4046 Units',
               '4225 Units','4770 Units','TotalBagged Units',
               'SmlBagged Units','LrgBagged Units','X-LrgBagged Units']]
```

```python
final.sort_index(inplace=True)
final
```

## Group By and Plots

```python
reg = final.groupby(['Regions'])
```

```python
USA = reg.get_group('United States')
```

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

plt.style.use('seaborn')

# Create figure and plot space
fig, ax = plt.subplots(figsize=(10, 10))

# Add x-axis and y-axis
ax.plot(USA.index.values,
        USA['Mean Price'], color='green', linewidth=2, label='All Avocados')

ax.plot(USA.index.values,
         USA['Regular Price'], color='red', linewidth=1, label='Regular Price')

ax.plot(USA.index.values,
         USA['Organic Price'], color='blue', linewidth=1, label='Organic Price')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Price",
       title="USA Price")

ax.legend()

plt.tight_layout()
plt.show()
```
