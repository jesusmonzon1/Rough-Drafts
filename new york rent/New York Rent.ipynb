## NYC 2 Bedroom Rent
A friend wanted to compare two neighberhoods. Figured this would be a quick exercise in practicing my python skills.
Data: https://streeteasy.com/blog/download-data/

```python
import pandas as pd
import numpy as np
import datetime as dt

df = pd.read_csv('medianTwoBedroomRent.csv')
```

```python
adam = ['Crown Heights', 'Astoria', 'All Queens', 'All Brooklyn']
filt = df['areaName'].isin(adam)
ad2 = df.loc[filt]
```

```python
ad2 = ad2.T
```

```python
list(ad2.columns.values)
```

```python
ad2.rename(columns={6:'Astoria'}, inplace=True)
ad2.rename(columns={53:'Crown Heights'}, inplace=True)
```

```python
ad2.index.name = "Date"
```

```python
ad2.drop(['areaName', 'Borough', 'areaType'], inplace=True)
```

```python
ad2['Years'] = ad2.index

ad2['Years'] = pd.to_datetime(ad2['Years'], format= '%Y-%m')

ad2.set_index('Years', inplace=True)

# d2.drop(columns=['Years'], inplace=True)
```

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

plt.style.use('seaborn')

# Create figure and plot space
fig, ax = plt.subplots(figsize=(10, 10))

# Add two lines
ax.plot(ad2.index.values,
        ad2['Astoria'], color='red', linewidth=2, label='Astoria')

ax.plot(ad2.index.values,
         ad2['Crown Heights'], color='blue', linewidth=2, label='Crown Heights')


# Set title and labels for axes
ax.set(xlabel="Years",
       ylabel="Per Month Rent",
       title="Rent Comparison For 2 Bedroom")

ax.legend()

plt.tight_layout()
plt.show()
```
